```python
def blosum62(sequence):
    '''
    Argument: seq, a protein or a peptide sequence
    Return: blosum62 Matrix
    '''
    aminoAcidSymbols = 'A  C  D  E  F  G  H  I  K  L  M  N  P  Q  R  S  T  V  W  Y'.split()
    
    Feature = []
    for aminoAcid in sequence: # Read each AA
        feature = []           # Store each row
        V = aminoAcidSymbols
        V = itertools.product(aminoAcid, V)
        V = [''.join(i) for i in V]
        V = list(V)
        
        for v in V:
            feature.append(table[v])
        Feature.append(feature)
        
    Feature = np.array(Feature)
    return Feature
#end-def
```

```python
def readFASTAs(fileName):

    '''
    :param fileName:
    :return: genome sequences
    '''
    with open(fileName, 'r') as file:
        v = []
        genome = ''
        for line in file:
            if line[0] != '>':
                genome += line.strip()
            else:
                v.append(genome.upper())
                genome = ''
        v.append(genome.upper())
        del v[0]
        return v
```
